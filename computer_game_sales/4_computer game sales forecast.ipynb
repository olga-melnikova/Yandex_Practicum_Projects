```python
# функция для вывода на экран таблицы с процентов пропущенных значений
def null_sum(df=data_games, name_1='null_value'):
    isna_games = df.isna().sum().to_frame(name=name_1)
    isna_games['percent_null'] = round((isna_games['null_value'] / len(df) * 100), 2)
    return isna_games
```

**Оценки игр**

Рассмотрим столбцы с оценкой игр.  
Посмотрим есть ли зависимость появления пропусков в рейтингах от наименования игры/платформы/даты релиза/жанра.

```python
# проверим зависимость появления пропусков от других данных представленных в исходной таблице
# напишем функцию, которая группирует по категорям столбца column 
# и возвращает процент пропусков по категории

# cтолбец na_sales выбран для подсчета количества значений т.к. в нем нет пропусков и 
# он не будет использоваться для группировки по категориям в текущем разделе

def p_null(df=new_data_games, index_1='platform'):
    
    score_null = df.pivot_table(index=df[index_1], 
                                values=['na_sales', 'critic_score', 'user_score', 'rating'], 
                                aggfunc='count').reset_index()
    
    score_null['percent_null_critic'] = round((((1 - (score_null['critic_score'] / score_null['na_sales'])) * 100)), 0)
    
    score_null['percent_null_user_score'] = round((((1 - (score_null['user_score'] / score_null['na_sales'])) * 100)), 0)
    
    score_null['percent_null_raiting'] = round((((1 - (score_null['rating'] / score_null['na_sales'])) * 100)), 0)
    
    return score_null
```

```python
pp = p_null(df=new_data_games, index_1='platform')
pp.head()
```

Для наглядности построим график по полученной сводной таблице.

```python
plt.figure(figsize=(20, 5))
bar = sns.barplot(data=pp, x='platform', y='percent_null_critic', palette='mako')
bar.set_title('Процент пропущенных значений в столбце оценки критиков', fontdict={'fontsize': 12})
bar.set_xlabel('платформа', fontsize=12)
bar.set_ylabel('% пропусков', fontsize=12)
plt.show()
```

На платформах 2600, GB, GEN и еще многи нет ни одной оценки критика/пользотваеля/ESRB.

Проверим на графике так ли, что 100% пропущенных значений у устаревших платформ.

```python
py = p_null(df=new_data_games, index_1='year_of_release')
py.head()
```

Применительно для рейтинга игр аббревиатура "TBD" (To Be Determined) означает, что рейтинг еще не определен. Возможно на момент выгрузки данных рейтинги обновлялись и появилась аббревиатура "TBD". Так как мы не можем сказать какой рейтинг был присвоен в итоге, данное значение в данных можно воспринять так же как пропуски.  

В одном жанре игр всегда есть более удачные и менее удачные, при этом бывает, что высокая оценка пользователей не совпадает с оценкой критиков и наоборот. На разных платформах одна и та же игра может иметь разный успех у пользователей. ПОэтому заменять на среднее будет не совсем верно. Заменим в данных как пропуски так и TBD на оценку, которой в данных нет. 

```python
# какие есть значения в 'user_score' 
new_data_games['user_score'].unique()
```

Так как в оценках уже имеется 0 - заменить на него пропуски и TBD мы не можем, данные сольются.  
Заменим пропуски на -1 - для TBD и Nan.  

```python
# замена пропусков и TBD в столбце 'user_score'
new_data_games.loc[new_data_games['user_score'] == "TBD", 'user_score'] = -1
new_data_games['user_score'] = new_data_games['user_score'].fillna(-1)
```

Уникальные значения столбца rating. На сайте ESRB представлены следующие категории рейтингов:

```python
# построение столбчатого графика для оценки категориальных переменных
def bar_mako(df, x1, y1, title_1, xlabel_1, ylabel_1, legend_1):    
    plt.figure(figsize=(20, 5))
    sns.set_style("whitegrid")
    bar = sns.barplot(data=df, 
                      x=x1, y=y1, palette='mako')
    bar.set_title(title_1, fontdict={'fontsize': 14})

    bar.set_xlabel(xlabel_1, fontsize=14)
    bar.set_ylabel(ylabel_1, fontsize=14)
    bar.legend([legend_1], fontsize=14)
    return bar
```

```python
# построение линейного графика с различными категориями
def lineplot_mako(df, x1, y1, hue1, title_1, xlabel_1, ylabel_1, figsize1=(20, 5)):
    plt.figure(figsize=figsize1)
    line = sns.lineplot(data=df, 
                        x=x1, 
                        y=y1, 
                        hue=hue1, 
                        palette='mako',
                        marker='o', 
                        linewidth=3)
    line.set_title(title_1, fontdict={'fontsize': 14})

    line.set_xlabel(xlabel_1, fontsize=14)
    line.set_ylabel(ylabel_1, fontsize=14)

    return line 
```

```python
# график бар разделяющий категории по подкатегорям
def bar_hue(df, x1, title1, xt, yt, width1=0.8):
    
    cmap = plt.get_cmap('GnBu')
    b_colors = cmap(np.array([i for i in range(100, 300, 50)]))
    bar_hue = df.plot(kind='bar', x=x1, color=b_colors, figsize=(20, 5), width=width1)
    bar_hue.set_title(title1, fontdict={'fontsize': 14})
    bar_hue.set_xlabel(xt, fontsize=14)
    bar_hue.set_ylabel(yt, fontsize=14)

    return bar_hue
```

```python
# точечная диаграмма с разделением по категорям
def scatter_mako(df, x1, y1, hue1, title_1, xlabel_1, ylabel_1, figsize1=(20, 5)):
    plt.figure(figsize=figsize1)
    
    scatter_m = sns.scatterplot(data=df, 
                            x=x1, 
                            y=y1, 
                            alpha=0.7,
                            hue=hue1,
                            palette='mako',
                            linewidth=0)
    scatter_m.set_title(title_1, fontdict={'fontsize': 14})
    scatter_m.set_xlabel(xlabel_1, fontsize=14)
    scatter_m.set_ylabel(ylabel_1, fontsize=14)
    
    return scatter_m

```

```python
# бар для нескольких графиков на листе
def bar_mako_2(df, x1, y1, title_1, xlabel_1, ylabel_1, ax):    
    plt.figure(figsize=(7, 5))
    sns.set_style("whitegrid")
    bar = sns.barplot(data=df, 
                      x=x1, y=y1, palette='mako', ax=axes[ax])
    bar.set_title(title_1, fontdict={'fontsize': 14})

    bar.set_xlabel(xlabel_1, fontsize=14)
    bar.set_ylabel(ylabel_1, fontsize=14)

    
    return bar
```

```python
# оценим корреляцию столбца цены (полной)
# назначим размер тепловой карты
# создадим теплову карту по платформе xbox_360 по столбцам 'critic_score', 'user_score'

def heatmap_b(df, col=['critic_score', 'user_score'], col_1='critic_score', title_1='Корреляция'):
    plt.figure(figsize=(20, 5))

    heatmap_b = sns.heatmap((df.corr()[col].sort_values(by=col_1)), 
        vmin=-1, 
        vmax=1, 
        cmap="YlGnBu", 
        annot=True, 
        fmt='.2g')

    heatmap_b.set_title(title_1, fontdict={'fontsize':14})

    return heatmap_b
```

```python
# построение круговой диаграммы для оценки долей

def pie_plot(df, labels1):
    
    plt.figure(figsize=(20, 5))
    
    explode=(0.2, 0.2, 0.2, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3)
    
    offset=1.4
    
    cmap = plt.get_cmap('GnBu')
    b_colors = cmap(np.array([i for i in range(120, 360, 15)]))

    pie_plot = plt.pie(x=df, labels=labels1, 
                       shadow=True, labeldistance=1.1, 
                       startangle=180, radius=1.5, 
                       explode=explode, rotatelabels=True,
                       colors=b_colors, 
                       wedgeprops=dict(width=offset, edgecolor='w'), 
                       autopct='%0.1f%%')
    return pie_plot
```

```python
# сводная таблица по жанрам и продажам, в том числе регионов
# в данном разделе нужны только глобальные и количество выпущенных игр по жанрам

top_genre = actual_data_games.pivot_table(
    index='genre', 
    values=['name', 'na_sales', 'eu_sales', 
            'jp_sales', 'other_sales', 
            'total_sales'], 
    aggfunc={'name': 'count', 
             'na_sales': 'sum', 
             'eu_sales': 'sum',
             'jp_sales': 'sum', 
             'other_sales': 'sum', 
             'total_sales': 'sum'}).sort_values(by='total_sales', 
                                                ascending=False).reset_index()
top_genre = top_genre[['genre', 'name', 'eu_sales', 'jp_sales', 'na_sales',  'other_sales', 'total_sales']]
top_genre[['genre', 'name', 'total_sales']]
```

Выделим ТОП-10 самых продаваемых игр по годам с 2011 по 2016 и и посмотрим какие оценки получали в среднем эти игры от критиков и пользователей каждый год.

```python
#для последующей работы отфильтруем датасет, чтобы не попали значения замененных пропусков.

a = actual_data_games.query('critic_score > 0 and user_score >= 0').sort_values(by='total_sales', ascending=False)

# сохраним в списки год, среднюю оценку критиков и пользователей самых проаваемых 10 игр
# и стандартное отклонение по ним

year, mean_critic_score, std_critic, mean_user_score, std_user = [], [], [], [], []

for i in range(2011, 2017):
    a1 = a.query('year_of_release == @i').sort_values(by='total_sales', ascending=False).head(10)
    year.append(i)
    mean_critic_score.append(a1['critic_score'].mean())
    mean_user_score.append(a1['user_score'].mean())
    std_critic.append(a1['critic_score'].std())
    std_user.append(a1['user_score'].std())
    

# создадим датафрейм mean_rating с полученными данными

mean_rating = pd.DataFrame({'year': year, 
                           'mean_critic_score': mean_critic_score,
                           'std_critic': std_critic,
                           'mean_user_score': mean_user_score,
                           'std_user': std_user})

for col in mean_rating.columns:
    mean_rating[col]=np.around(mean_rating[col], 2)
mean_rating
```

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H0 - нулевая гипотеза: 
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.  


H1 - альтернативная гипотеза:
- Средние пользовательские рейтинги платформ Xbox One и PC различаются.  

Для проверки гипотезы о равенстве средних сохраним два датасета из общего актуального - один с платформой Xbox One, другой с PC и отфильтруем значения рейтинга, которые были заменены вручную. Выедем статистичесакую информацию по получившимся датасетам и посчитаем дисперсию. 

```python
pc = actual_data_games.query('platform == "PC" and user_score >= 0')
pc[['user_score']].describe().T
```

```python
# ddof=1 по умолчанию для выборки
print(f'Дисперсия пользовательских оценок игр для PC равна {pc["user_score"].var()}')
print(f'Стандартное отклонение пользовательских оценок игр для PC равно {pc["user_score"].std()}')
```

По величине стандартного отклонения пользовательских оценок на PC можно сказать, что оценки в целом, не сильно различаются для различных игр - стандартное отклонение ниже среднего. 

```python
xbone = actual_data_games.query('platform == "XONE" and user_score >= 0')
xbone[['user_score']].describe().T
```

```python
# ddof=1 по умолчанию для выборки
print(f'Дисперсия пользовательских оценок игр на платформе XBox ONE равна {xbone["user_score"].var()}')
print(f'Стандартное отклонение пользовательских оценок игр на платформе XBox ONE равно {xbone["user_score"].std()}')
```

Стандартное отклонение пользовательских оценок XBox ONE ниже - 1,38, чем у PC - 1,61, но и выборка по размеру меньше - 182 и 307.

Примем критический уровень статистической значимости alpha=0,05 (5%) - если p_value окажется меньше, то нулевая гипотеза отвергается.  
Две генеральные совокупности друг от друга независимы. Для подтверждения/опровержения гипотезы воспользуемся методом сравнения среднего двух генеральных совокупностей.  

```python
alpha = .05 # критический уровень статистичесой значимости

results = st.ttest_ind(pc['user_score'], xbone['user_score']) 

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

По выборке по актуальному периоду отвергнуть нулевую гипотезу, о том, что средние пользовательские оценки двух патформ равны мы не можем. Средние оценки пользователей на XBox ONE и PC в актуальном периоде практически равны.

```python
plt.figure(figsize=(20, 5))
a = sns.kdeplot(data=pc['user_score'], shade=True, legend=True, color='mediumblue')
a = sns.kdeplot(data=xbone['user_score'], shade=True, legend=True, color='lightgreen')
a.axes.set_title('График плотности распределения оценок пользователей игр жанров Action и Sports', fontsize=12)
a.set_xlabel('Оценки пользователей', fontsize=12)
a.legend(['PC', 'XBOX ONE'], loc=1, prop={'size': 12})
plt.show()
```

**Вывод**
***

Гипотеза о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые подтвердилась. По выборке мы точно не можем сказать, что средние рейтинги отличаются.

- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») **равны**.

H1 - альтернативная гипотеза:

- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») **разные**.

Для проверки гипотезы о равенстве средних сохраним два датасета из общего актуального - один с жанром Action, другой с жанром Sports и отфильтруем значения рейтинга, которые были заменены вручную. Выедем статистичесакую информацию по получившимся датасетам и посчитаем дисперсию. 

```python
action = actual_data_games.query('genre == "ACTION" and user_score >= 0')
action[['user_score']].describe().T
```

```python
# ddof=1 по умолчанию для выборки
print(f'Дисперсия пользовательских оценок игр жанра Action равна {action["user_score"].var()}')
print(f'Стандартное отклонение пользовательских оценок игр жанра Action равно {action["user_score"].std()}')
```

```python
sports = actual_data_games.query('genre == "SPORTS" and user_score >= 0')
sports[['user_score']].describe().T
```

```python
# ddof=1 по умолчанию для выборки
print(f'Дисперсия пользовательских оценок игр жанра Sports равна {sports["user_score"].var()}')
print(f'Стандартное отклонение пользовательских оценок игр жанра Sports равно {sports["user_score"].std()}')
```

Выборки не равны.  

Примем критический уровень статистической значимости alpha=0,05 (5%) - если p_value окажется меньше, то нулевая гипотеза отвергается.  
Две генеральные совокупности друг от друга независимы. Для подтверждения/опровержения гипотезы воспользуемся методом сравнения среднего двух генеральных совокупностей.  

```python
alpha = .05 # критический уровень статистичесой значимости

results = st.ttest_ind(sports['user_score'], action['user_score']) 

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Нулевая гипотеза не подтвердилась, принимаем альтернативную гипотезу: **средние пользовательствие оценки жанров Action и Sports разные.**

```python
plt.figure(figsize=(20, 5))
a = sns.kdeplot(data=sports['user_score'], shade=True, legend=True, color='mediumblue')
a = sns.kdeplot(data=action['user_score'], shade=True, legend=True, color='lightgreen')
a.axes.set_title('График плотности распределения оценок пользователей игр жанров Action и Sports', fontsize=12)
a.set_xlabel('Оценки пользователей', fontsize=12)
a.legend(['SPORTS', 'ACTION'], loc=1, prop={'size': 12})
plt.show()
```

**Вывод**
***  
Принята альтернативная гипотеза - средние пользовательствие оценки жанров Action и Sports разные.
