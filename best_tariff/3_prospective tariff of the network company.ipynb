Клиентам компании «Мегалайн» — федерального оператора сотовой связи предлагают два тарифных плана: «Смарт» и «Ультра».  

Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.
Необходимо провести предварительный анализ тарифов на небольшой выборке клиентов. В нашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.  

__Описание тарифов__  
**`Тариф «Смарт»`**  
- Ежемесячная плата: 550 рублей  
- Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика  
- Стоимость услуг сверх тарифного пакета:   
    1. минута разговора: 3 рубля («Мегалайн» всегда округляет вверх значения минут и мегабайтов. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута);   
    2. сообщение: 3 рубля;   
    3. 1 Гб интернет-трафика: 200 рублей. 
    
    
**`Тариф «Ультра»`**  

- Ежемесячная плата: 1950 рублей  
- Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
- Стоимость услуг сверх тарифного пакета:  
    1. минута разговора: 1 рубль;  
    2. сообщение: 1 рубль;   
    3. 1 Гб интернет-трафика: 150 рублей.  

***Примечание:***  
«Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт:  
- Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.  
- Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.


Мегалайн округляет до гб в месяц, поэтому в таблице округлим суммарное количество использованного трафика сверх тарифа вверх, а включенное учтем по таблице.  
В датасет добавим столбцы:  
1. Количество минут сверх тарифа - 'call_over'
2. Количество смс сверх тарифа - 'sms_over'
3. Количество интеренет траффика сверх тарифа, округленное вверх до Гб - 'gb_over'

```python
data_clients['call_over'] = data_clients['sum_duration'] - data_clients['minutes_included']
data_clients['sms_over'] = data_clients['count_sms'] - data_clients['messages_included']
data_clients['gb_over'] = np.ceil((data_clients['sum_mb'] - data_clients['mb_per_month_included']) / 1024)
```

```python
over_list = ['call_over', 'sms_over', 'gb_over']
for over in over_list:
    data_clients[over] = np.maximum(data_clients[over], 0)
```

Добавляем в таблицу итоговый доход по каждому пользователю в месяц:

- 'sms_over_income' - смс сверх тарифа на цену за доп. смс
- 'internet_over_income' - гб сверх тарифа, округленные вверх до гб умножаем на цену за гб сверх тарифа

Выводим итоговую выручку с пользователя в каждом месяце в столбец 'total_income'.

```python
data_clients['call_over_income'] = data_clients['call_over'] * data_clients['rub_per_minute']
data_clients['sms_over_income'] = data_clients['sms_over'] * data_clients['rub_per_message']
data_clients['internet_over_income'] = data_clients['gb_over'] * data_clients['rub_per_gb']

# общий доход от тарифа по пользователю в месяц (тарифная ставка в месяц + дополнительные услуги)
data_clients['total_income'] = (
    data_clients['rub_monthly_fee'] + 
    data_clients['call_over_income'] + 
    data_clients['sms_over_income'] + 
    data_clients['internet_over_income'])
```

### Проверка гипотез

### Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

H0 - нулевая гипотеза: 
- Средние выручки пользователей тарифов «Ультра» и «Смарт» равны.  


H1 - альтернативная гипотеза:
- Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются.  

Примем критический уровень статистической значимости alpha=0,05 (5%) - если p_value окажется меньше, то нулевая гипотеза отвергается.  
Две генеральные совокупности друг от друга независимы. Для подтверждения/опровержения гипотезы воспользуемся методом сравнения среднего двух генеральных совокупностей.

```python
smart_income = smart_clients['total_income']
ultra_income = ultra_clients['total_income']
```

```python
alpha = .05 # критический уровень статистичесой значимости

results = st.ttest_ind(smart_income, ultra_income) 

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

```python
print('Средняя выручка от пользователей тарифа Smart', round((smart_income.mean()), 0))
print('Средняя выручка от пользователей тарифа Ultra', round((ultra_income.mean()), 0))
```

**Вывод**
***
Гипотеза о том, что выручка на тарифах различается - верна: средняя выручка от пользователей тарифа Smart - 1303.0, тарифа Ultra - 2072.0. 

### Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

H0 - нулевая гипотеза: 
- Средняя выручка пользователей из Москвы равна средней выручке пользователей из других регионов.  


H1 - альтернативная гипотеза:
- Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.  

Примем критический уровень статистической значимости alpha=0,05 (5%) - если p_value окажется меньше, то нулевая гипотеза отвергается.

```python
data_clients.head()
```

```python
moscow_income = data_clients.query('city == "Москва"')['total_income']
else_income = data_clients.query('city != "Москва"')['total_income']
```

```python
alpha = .05 # критический уровень статистичесой значимости

results = st.ttest_ind(moscow_income, else_income) 

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

```python
print('Средняя выручка от пользователей по Москве', round((moscow_income.mean()), 0))
print('Средняя выручка от пользователей по другим городам', round((else_income.mean()), 0))
```
